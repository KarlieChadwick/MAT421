**Chapter 21.1: Numerical Integration**

The approximation of an integral over a given period ([a,b]) can be found by determining the area under a given function, f(x). There are many different methods used to compute the area that will be discussed in the next two chapters. A big part of these methods use subintervals of the area which are just smaller shapes between $[x_i, x_{i+1}]$ within the total interval.

**Chapter 21.2: Riemanns Integral**

Reimanns sum is a method that can be used to approximate integrals by summing a group of rectangular subintervals. The width of each rectangle are equal in value and can be found using the equation below;

$h = x_{i+1}-x_i$ 

The height is determined by the function itself. Because it is a rectangular area, the height can be determined as either the value at the left endpoint, midpoint, or right endpoint. 

*Integral Equations:*

* Left Endpoint - $∫_a^bf(x)dx≈∑_{i=0}^{n−1}hf(x_i)$

* Right Endpoint - $∫_a^bf(x)dx≈∑_{i=1}^nhf(x_i)$

* Midpoint - $∫_a^bf(x)dx≈∑_{i=0}^{n−1}hf(y_i)$, where $y_i = \frac{x_{i+1}+x_i}{2}$

*1) Use each Riemann Integral Equation to approximate the integral of x^2 between 0 and 10 with 6 evenly spaced grid points.*

```python
import numpy as np

# The constraints given in the question are defined.
# This included the equation, interval, and step size.

a = 0
b = 10
n = 6
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = x**2

# Each of the three equations are defined and used here
# to calculate the sums.

Left_Riemann = h * sum(f[:n-1])

Right_Riemann = h * sum(f[1::])

Mid_Riemann = h * sum(((x[:n-1] \
        + x[1:])/2)**2)

# The values are returned.

print('The value found using left endpoints is ' + str(Left_Riemann))

print('The value found using right endpoints is ' + str(Right_Riemann))

print('The value found using midpoints is ' + str(Mid_Riemann))

# A graph of the equation is printed with the area
# under the curved filled in for a visual of the problem.

import matplotlib.pyplot as plt

fig = plt.figure(figsize = (9,6))
plt.title('Area under the Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color = 'grey')

t = np.linspace(-8,13)
plt.plot(t,t**2,'r')
plt.fill_between(t,t**2, where = (0<=t)&(t<=10), color = 'red', alpha = 0.3)

plt.show()
```

*2) Use each Riemann Integral Equation to approximate the integral of cos(x) between 0 and pi/2 with 11 evenly spaced grid points.*

```python
import numpy as np

a = 0
b = (np.pi)/2
n = 11
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

Left_Riemann = h * sum(f[:n-1])

Right_Riemann = h * sum(f[1::])

Mid_Riemann = h * sum(np.cos((x[:n-1] \
        + x[1:])/2))

print('The value found using left endpoints is ' + str(Left_Riemann))

print('The value found using right endpoints is ' + str(Right_Riemann))

print('The value found using midpoints is ' + str(Mid_Riemann))

import matplotlib.pyplot as plt

fig = plt.figure(figsize = (9,6))
plt.title('Area under the Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color = 'grey')

t = np.linspace(0,(2*np.pi))
plt.plot(t,np.cos(t),'r')
plt.fill_between(t,np.cos(t), where = (0<=t)&(t<=(np.pi/2)), color = 'red', alpha = 0.3)

plt.show()
```

*3) Compare the error between the actual value and approximated value of the integral $x^2+3$ between 2 and 10 with 3 subintervals and 9 subintervals using the midpoint rule.*



```python
import numpy as np

a = 2
b = 10

n_1 = 3
h_1 = (b - a) / (n_1 - 1)
x_1 = np.linspace(a, b, n_1)
f = x_1**2+3

Mid_Riemann_1 = h_1 * sum(((x_1[:n_1-1] \
        + x_1[1:])/2)**2+3)
error_1 = round((((1064/3) - Mid_Riemann_1)/(1064/3))*100,2)

print('The value when 3 subintervals were used was ' + str(Mid_Riemann_1))
print('The percent error between the approximated and actual value is ' + str(error_1)+'%')

n_2 = 9
h_2 = (b - a) / (n_2 - 1)
x_2 = np.linspace(a, b, n_2)
f = x_2**2+3

Mid_Riemann_2 = h_2 * sum(((x_2[:n_2-1] \
        + x_2[1:])/2)**2+3)
error_2 = round((((1064/3) - Mid_Riemann_2)/(1064/3))*100,2)

print('The value when 9 subintervals were used was ' + str(Mid_Riemann_2))
print('The percent error between the approximated and actual value is ' + str(error_2)+'%')

import matplotlib.pyplot as plt

fig = plt.figure(figsize = (9,6))
plt.title('Area under the Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color = 'grey')

t = np.linspace(1,11)
plt.plot(t,t**2+3,'r')
plt.fill_between(t,t**2+3, where = (2<=t)&(t<=10), color = 'red', alpha = 0.3)

plt.show()
```

From the above answer we can see that using a larger amount of subintervals will create a smaller error between the approximated and exact integral value.

**Chapter 21.3: Trapezoid Rule**

The trapezoid rule is simmilar to the Riemann rule, however instead of intervals of rectangles, trapezoids are use to get a more acurate read of the curve. The approximation equation can be written as: $∫_a^bf(x)dx≈∑_{i=0}^{n−1}h\frac{f(x_i)+f(x_{i+1})}{2}$

*1) Find the approximation and percent error of the integral of cos(x) given the interval [0,pi/2] and 11 equal gridpoints.*

```python
import numpy as np

a = 0
b = (np.pi)/2
n = 11
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

Trapezoid = (h/2)*(f[0] + \
          2 * sum(f[1:n-1]) + f[n-1])
error = round((((1 - Trapezoid)/1)*100),2)

print('The value using the trapezoid rule is ' + str(Trapezoid))
print('The percent error between the approximated and exact value is ' + str(error)+'%')
```

*2) Find the approximation of the integral of x^2+4x-3 given the interval [0,5] and 6 equal gridpoints.*

```python
import numpy as np

a = 0
b = 5
n = 6
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = x**2+4*x-3

Trapezoid = (h/2)*(f[0] + \
          2 * sum(f[1:n-1]) + f[n-1])

print('The value using the trapezoid rule is ' + str(Trapezoid))

import matplotlib.pyplot as plt

fig = plt.figure(figsize = (9,6))
plt.title('Area under the Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color = 'grey')

t = np.linspace(-2,7)
plt.plot(t,t**2+4*t-3,'r')
plt.fill_between(t,t**2+4*t-3, where = (0<=t)&(t<=5), color = 'red', alpha = 0.3)

plt.show()
```
