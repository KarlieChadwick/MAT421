**17.4: Lagrange Polynomial Interpolation**

Ih chapter 17.3 we found a group of polynomial equations that was able to pass through the set of data points given. With lagrange polynomial interpolation, the computer will find a single polynomial equation that satisfies the given data.

To start calculations for the lagrange polynomial, the equation below is used to first find lagrange basis polynomials:

$ùëÉ_ùëñ(ùë•)=‚àè^ùëõ_{j=1,ùëó‚â†ùëñ}(ùë•‚àíùë•_ùëó)/(ùë•_ùëñ‚àíùë•_ùëó)$

After the basis polynomials are found, they are put into this equation to find the single lagrange polynomial:

$ùêø(ùë•)=‚àë^ùëõ_{ùëñ=1}ùë¶_ùëñùëÉ_ùëñ(ùë•)$

*1) Find the Lagrange basis polynomials for the data set x = [1,2,3] and y = [3,5,2] then graph the lagrange polynomial interpolation*



- $P_1(x) = \frac{((x-2)(x-3))}{((1-2)(1-3))} = \frac{1}{2}x^2-\frac{5}{2}x+3$


- $P_2(x) = \frac{((x-1)(x-3))}{((2-1)(2-3))} = -x^2+4x-3$


- $P_3(x) = \frac{((x-1)(x-2))}{((3-1)(3-2))} = \frac{1}{2}x^2-\frac{3}{2}x+1$

```python
import numpy as np
import numpy.polynomial.polynomial as poly

x = [1, 2, 3]
y = [3, 5, 2]

# The coefficents are found using the basis polynomial
# equations above. The first coefficient listed is the
# C value of the polynomial and so on.

P1_coeff = [3,-2.5,.5]
P2_coeff = [-3, 4,-1]
P3_coeff = [1,-1.5,.5]

# The poly funciton imported above will compute a polynomial
# funciton for each basis polynomial to be graphed below.

P1 = poly.Polynomial(P1_coeff)
P2 = poly.Polynomial(P2_coeff)
P3 = poly.Polynomial(P3_coeff)

x_new = np.arange(-1.0, 4.0, 0.1)

# The plt function is used to plot the ploynomials as well as
# format the graph colors and labels.

import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

fig = plt.figure(figsize = (10,6))

plt.plot(x_new, P1(x_new), 'b', label = 'P1')
plt.plot(x_new, P2(x_new), 'r', label = 'P2')
plt.plot(x_new, P3(x_new), 'g', label = 'P3')
plt.plot(x, np.ones(len(x)), 'ko', x, np.zeros(len(x)), 'ko')

plt.title('Lagrange Basis Polynomials')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color = 'white')
plt.legend(frameon = True, shadow = True)
plt.show()
```

```python
# After the basis polynomials are found, the second
# equation is employed to find the single lagrange
# polynomial. This equation esentially adds the basis
# equations together multiplied at each given y 
# counterpoint.

L = 3*P1 + 5*P2 + 2*P3

fig = plt.figure(figsize = (10,6))

plt.plot(x_new, L(x_new), 'k', x, y, 'ro')

plt.title('Lagrange Polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color = 'white')
plt.show()
```

*2) Find the lagrange basis polynomial for the dataset x = [0,1,2,3,4] and y = [4,3,7,2,1] then graph the lagrange polynomial interpolation*

- $P_1(x) = \frac{(x-1)(x-2)(x-3)(x-4)}{(0-1)(0-2)(0-3)(0-4)} = \frac{1}{24}x^4-\frac{5}{12}x^3+\frac{35}{24}x^2-\frac{25}{12}x+1$


- $P_2(x) = \frac{(x-0)(x-2)(x-3)(x-4)}{(1-0)(1-2)(1-3)(1-4)} = -\frac{1}{6}x^4+\frac{3}{2}x^3-\frac{13}{3}x^2+4x+0$


- $P_3(x) = \frac{(x-0)(x-1)(x-3)(x-4)}{(2-0)(2-1)(2-3)(2-4)} = \frac{1}{4}x^4-2x^3+\frac{19}{4}x^2-3x+0$


- $P_4(x) = \frac{(x-0)(x-1)(x-2)(x-4)}{(3-0)(3-1)(3-2)(3-4)} = -\frac{1}{6}x^4+\frac{7}{6}x^3-\frac{7}{3}x^2+\frac{4}{3}x+0$


- $P_5(x) = \frac{(x-0)(x-1)(x-2)(x-3)}{(4-0)(4-1)(4-2)(4-3)} = \frac{1}{24}x^4-\frac{1}{4}x^3+\frac{11}{24}x^2-\frac{1}{4}x+0$

```python
import numpy as np
import numpy.polynomial.polynomial as poly

x = [0, 1, 2, 3, 4]
y = [4, 3, 7, 2, 1]

P1_coeff = [1,-25/12,35/24,-5/12,1/24]
P2_coeff = [0,4,-13/3,3/2,-1/6]
P3_coeff = [0,-3,19/4,-2,1/4]
P4_coeff = [0,4/3,-7/3,7/6,-1/6]
P5_coeff = [0,-1/4,11/24,-1/4,1/24]

P1 = poly.Polynomial(P1_coeff)
P2 = poly.Polynomial(P2_coeff)
P3 = poly.Polynomial(P3_coeff)
P4 = poly.Polynomial(P4_coeff)
P5 = poly.Polynomial(P5_coeff)

x_new = np.arange(-1.0, 5.0, 0.1)

import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

fig = plt.figure(figsize = (10,6))

plt.plot(x_new, P1(x_new), 'b', label = 'P1')
plt.plot(x_new, P2(x_new), 'r', label = 'P2')
plt.plot(x_new, P3(x_new), 'g', label = 'P3')
plt.plot(x_new, P4(x_new), 'orange', label = 'P4')
plt.plot(x_new, P5(x_new), 'blueviolet', label = 'P5')
plt.plot(x, np.ones(len(x)), 'ko', x, np.zeros(len(x)), 'ko')

plt.title('Lagrange Basis Polynomials')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color = 'white')
plt.legend(frameon = True, shadow = True)
plt.show()
```

```python
L = 4*P1 + 3*P2 + 7*P3 + 2*P4 + 1*P5

fig = plt.figure(figsize = (10,4))

plt.plot(x_new, L(x_new), 'hotpink', x, y, 'bo')

plt.title('Lagrange Polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color = 'white')
plt.show()
```

**17.5: Newton's Polynomial Interpolation**

Newton's polynomial interpolation can be used to fit a given data set to a polynomial. It can be written as the following equation where the polynomial goes through $n$ points:

$f(x)=\sum_{i=0}^na_in_i(x)$

The coefficients for the polynomial can be found using a technique called divided differences. The iteration can be written as:

$f[x_k,x_{k-1},...,x_1,x_0]=\frac{f[x_k,x_{k-1},...,x_2,x_2]-f[x_{k-1},x_{k-2},...x_1,x_0]}{x_k-x_0}$

*1)Create a function that will take a data set and solve a Newton's polynomial*

```python
import numpy as np

# The first defined function, divided_diff calculate the coefficents
# of the polynomial using the same iteration listed above. The for loop will
# run until there are no more variable coefficients to be found. The for loop
# in this function will allow for different length polynomials to be inputed
# without needing to change any of the code. 

def divided_diff(x, y):
    n = len(y)
    coef = np.zeros([n, n])
    coef[:,0] = y
    
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = \
           (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
            
    return coef

# The second defined function, Newton_poly will employ the above
# coefficients to create the actual Newton polynomial. The for loop
# here will run until the function has no more x data to account for.

def newton_poly(coef, x_data, x):
    n = len(x_data) - 1 
    p = coef[n]
    for k in range(1,n+1):
        p = coef[n-k] + (x -x_data[n-k])*p
    return p
```

*2) Find Newton's polynomial for the data set x = [3, 5, 2] and y = [-2, 1, 4] using defined function and then graph*

```python
%matplotlib inline

x = np.array([3, 5, 2])
y = np.array([-2, 1, 4])

# The funciton defined above is used to find the coefficients of the
# given data set and then the function is evaluated with new data points.

a_s = divided_diff(x, y)[0, :]

x_new = np.arange(min(x)-1, max(x)+1, .1)
y_new = newton_poly(a_s, x, x_new)

# The imported plt function is used to choose the style, dimension, labels
# and colors of the graph.

import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

plt.figure(figsize = (12, 6))

plt.plot(x_new, y_new, 'b')
plt.scatter(x, y, color='b',marker='o')
plt.title("Newton's Polynomial")
plt.xlabel('x')
plt.ylabel('y')
```

*3) Find Newton's polynomial for the data set x = [3, 12, 1, 0, 5, -2, -10] and y = [-2, 6, 4, 5, -7, 2, 1] using the defined function and then graph*

```python
%matplotlib inline

x = np.array([3, 12, 1, 0, 5, -2, -10])
y = np.array([-2, 6, 4, 5, -7, 2, 1])

a_s = divided_diff(x, y)[0, :]

x_new = np.arange(min(x)-1, max(x)+1, .1)
y_new = newton_poly(a_s, x, x_new)

import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

plt.figure(figsize = (12, 6))

plt.plot(x_new, y_new, 'g')
plt.scatter(x, y, color='g',marker='o')
plt.title("Newton's Polynomial")
plt.xlabel('x')
plt.ylabel('y')
```
