**Chapter 21.1: ODE Initial Value Problem Statement**

An ordinary differential equation or ODE is a differential equation in one dimension. It can be written as an incorporation of all arguments; a function, its independent variable, and any of its derivatives. 

A partial differential equation or PDE is a differential equation where any partial derivatives can be of any order. 

To solve a function, a general or particular solution can be determined. A general solution is a function that satisfies the problem. A particual solution is a function that satisfies the equation and the known values of the solution.

**Chapter 22.2: Reduction of Order**

In order to solve higher order equations the order can be reduced to a first order differential equation. To do so, the vector or state of the equation is set equal to the ODE matrix. Then, the state is derived. Once it is rewritten, it can be titled a linear ODE.

*1) Rewrite a first order ODE of the two equations below:*

- $\frac{da(t)}{dt} = 3a(t) + b(t)$

- $\frac{db(t)}{dt} = a(t) - 5b(t)$

$S(t) = [a(t), b(t)]$

$\frac{dS(t)}{dt} = [3, 1, 1, -5]S(t)$

The first matrix is 1X2 and the second matrix is 2X2.

**Chapter 22.3: The Euler Method**

To linearly approximate S(t) around $t_j$ at $t_{j+1}$ we use the Explicit Euler formula;
- $S(t_{j+1}) = S(t_j) + (t_{j+1} - t_j)\frac{dS(t_j)}{dt}$

When graphing the approximation, the step is an important factor. the small the step between values, the higher the accuracy of the approximation compared to the actual function. Here is an example with the equation $\frac{df(t)}{dt} = e^t$ with the initial condition $f_0 = 1$ when graphed with a step of 0.1 and a step of 0.05.

```python
import numpy as np
import matplotlib.pyplot as plt

# First, the parameters are defined including
# the ODE equation, step size, grid, initial
# condition

f = lambda t, s: np.exp(t)
h = 0.1 # Step size
t = np.arange(0, 1 + h, h)
s0 = 1

# Next, the equation is inputed in to the
# explicit Euler method

s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

# Lastly, the graphing parameters are set
# and the function is graphed against the 
# real solution

plt.style.use('ggplot')
%matplotlib inline

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, np.exp(t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for an ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid(color = 'grey')
plt.legend(loc='best')
plt.show()
```
