**17.1: Introduction to Interpolation**

Interpolation use a set of reliable data points to estimate unknown values that lie in between the known data points. There are two different types of interpolation discussed in the first part of this chapter; Linear Interpolation and Cubic Spline Interpolation



---



**17.2: Linear Interpolation**

Linear interpolation estimates an unknown value to lie on a line joining two known data points. The equation given to determine this value is:
$y(x) = y_{i} + ((y_{i + 1} - y_{i})(x - x_{i})) / (x_{i+1} - x_{i})$

*1) Find the linear interpolation at x = 2.5 based on the data x = [0, 2, 4] & y = [1, 5, 3] and plot.*

```python
from scipy.interpolate import interp1d

# I imported the interpolate library to Python. The data points of x & y were 
# given and then use in the interp1d function. The x point was inputed into
# the function to spit out the y counterpart that allowed for the point to fit
# on the linear data set.

x = [0, 2, 4]
y = [1, 5, 3]

f = interp1d(x, y)
y_hat = f(2.5)

print('The value of y at x = 2.5 is ' + str(y_hat))
```

```python
import matplotlib.pyplot as plt

# After importing the plt function in Python, I had fun determining the style
# of plotting I wanted to use as well as the colors of each line and point.

plt.style.use('ggplot')

plt.figure(figsize = (8,6)) # Determines the plot size
plt.plot(x,y,'-og')         # Plots the known values of x and y with a green line
plt.plot(2.5, y_hat, 'or')  # Marks the unknown value in red according to the
                            # y_hat value found above


# These functions allowed me to title and label the axis of the plot:

plt.title('Linear Interpolation at x = 2.5') 
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

*2) Find the linear interpolation at a = 3.5 based on the data a = [1,3,4,7,8,9] & b = [0,6,2,9,8,6] and plot.*

```python
from scipy.interpolate import interp1d

a = [1, 3, 4, 7, 8, 9]
b = [0, 6, 2, 9, 8, 6]

f = interp1d(a, b)
b_hat = f(3.5)

print('The value of b at a = 3.5 is ' + str(b_hat))
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')

plt.figure(figsize = (12,6)) 
plt.plot(a,b,'-or')         
plt.plot(3.5, b_hat, 'ob')

plt.title('Linear Interpolation at a = 3.5') 
plt.xlabel('a')
plt.ylabel('b')

plt.show()
```

* It is important to note that the unkown value 
must lie within the known data set for interpolation to occur! In the above example, a value of a = 10 would be out of range for the computer to process.

* The data set of y and x must also be equal in length so that each coordinate can be plotted. For example, x = [1, 5, 8, 9] & y = [2, 7, 4] would not be processed by the computer.

**17.3: Cubic Spline Interpolation**

A cubic spline interpolation is similar to the linear interpolation in that it is used to determine an unknown point within a given data set, however here, the function is a set of piecewise cubic functions rather than linear functions.

*1) Use CubicSpline to plot the cubic spline interpolation of the data set x = [1, 2, 3] and y = [0, 6, 3] for 1  x  3.*

```python
from scipy.interpolate import CubicSpline
import numpy as np

# After importing the CubicSpline and np functions, the given data set was 
# put into the CubicSpline function with the bc type as natural which assumes
# the second derivatives are 0 at the endpoints. The np.linespace function 
# determines the range for x which I've made to be the minimum and maximum
#  values in the x data set.

x = [1, 2, 3]
y = [0, 6, 3]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)
```

```python
import matplotlib.pyplot as plt

# Using the same plt function as in the linear interpolation graphs, I defined
# a plot style I liked as well as the graph dimensions, colors, and labels.
# I then used the variables defined above to graph the cubic spline interpolation.

plt.style.use('ggplot')

plt.figure(figsize = (8,6))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')

plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

* It is important to note that for cubic spline interpolation, the x data set must be stritctly increasing for the computer to process

*2) Use CubicSpline to plot the cubic spline interpolation of the data set a = [0, 5, 6, 8, 10, 25] and b = [1, 2, 5, 3 ,9, 4] for 0  a  25.*

```python
from scipy.interpolate import CubicSpline
import numpy as np

a = [0, 5, 6, 8, 10, 25]
b = [1, 2, 5, 3, 9, 4]

f = CubicSpline(a, b, bc_type='natural')
a_new = np.linspace(min(a), max(a), 100)
b_new = f(a_new)
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')

plt.figure(figsize = (14,8))
plt.plot(a_new, b_new, 'g')
plt.plot(a, b, 'ro')

plt.title('Cubic Spline Interpolation')
plt.xlabel('a')
plt.ylabel('b')

plt.show()
```
