# First Method: Cubic Spline Interpolation


## Part 1: Normal Graph, All Adults

```python
from scipy.interpolate import CubicSpline
import numpy as np

year = [2000, 2001, 2002, 2003, 2004, 2005,\
     2006, 2007, 2008, 2009, 2010, 2011,\
     2012, 2013, 2014, 2015, 2016, 2017, 2018]
Adults = [11863, 13006, 13391, 14012, 15126, 16186,\
     17110, 17273, 18651, 20490, 20974, 20589,\
     21319, 22173, 21872, 23161, 23104, 23048, 25158]

f = CubicSpline(year, Adults, bc_type='natural')
year_new = np.linspace(min(year), max(year), 100)
Adults_new = f(year_new)

import matplotlib.pyplot as plt

plt.style.use('ggplot')

plt.figure(figsize = (18,8))
plt.grid(color = 'white', linestyle = 'dashed')

plt.plot(year_new, Adults_new, 'b', label = 'Diabetic Adults (18+)')
plt.plot(year, Adults, 'black', label = 'Linear Data')
plt.plot(year, Adults, 'bo')

plt.legend(loc='upper left')

plt.title('Total Diabetic Adults - Cubic Spline Interpolation')
plt.ylabel('Number of Diabetic Adults (18+) in the US - In 1000s')
plt.xlabel('Year')

z = np.arange(2000,2019, step=(1))
plt.xticks(z, ['2000','2001','2002','2003','2004','2005'\
               ,'2006','2007','2008','2009','2010','2011'\
               ,'2012','2013','2014','2015','2016','2017','2018'])

plt.show()
```

# Second Method: Newtons Polynomial

## Part 1: Normal Graph, All Adults

```python
import numpy as np

def divided_diff(x, y):
    n = len(y)
    coef = np.zeros([n, n])
    coef[:,0] = y
    
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = \
           (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
            
    return coef

def newton_poly(coef, x_data, x):
    n = len(x_data) - 1 
    p = coef[n]
    for k in range(1,n+1):
        p = coef[n-k] + (x -x_data[n-k])*p

    return p
```

```python
year = [2000, 2001, 2002, 2003, 2004, 2005,\
     2006, 2007, 2008, 2009, 2010, 2011,\
     2012, 2013, 2014, 2015, 2016, 2017, 2018]

Adults = [11863, 13006, 13391, 14012, 15126, 16186,\
     17110, 17273, 18651, 20490, 20974, 20589,\
     21319, 22173, 21872, 23161, 23104, 23048, 25158]

a_s = divided_diff(year, Adults)[0, :]

year_new = np.arange(min(year)-1, max(year)+1, .1)
Adults_new = newton_poly(a_s, year, year_new)

import matplotlib.pyplot as plt

plt.figure(figsize = (18, 8))
plt.ylim([0, 30000])

plt.style.use('ggplot')

plt.grid(color = 'white', linestyle = 'dashed')

plt.plot(year, Adults, 'black', label = 'Linear Data')
plt.plot(year_new, Adults_new, 'b', label = 'Diabetic Adults (18+)')
plt.scatter(year, Adults, color='b',marker='o')

plt.legend(loc = 'upper left')

plt.title('Total Diabetic Adults - Newtons Polynomial Interpolation')
plt.ylabel('Number of Diabetic Adults (18+) in the US - In 1000s')
plt.xlabel('Year')

z = np.arange(1998,2020, step=(1))
plt.xticks(z, ['1998','1999','2000','2001','2002','2003','2004','2005'\
               ,'2006','2007','2008','2009','2010','2011'\
               ,'2012','2013','2014','2015','2016','2017','2018','2019'])

plt.show()
```
