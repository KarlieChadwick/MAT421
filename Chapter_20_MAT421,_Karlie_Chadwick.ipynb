**Chapter 20.1 - Numerical Differentiation Problem Statement**

A numerical grid can be defined as a set of points that are evenly spaced over a given interval. The spacing between the points is also known as the step size. For example, an interval from 1-5 with a step size of 1 permits the set {1, 2, 3, 4, 5}. 

Grids in one dimension can be generated via the linspace function in Python. A function f(x) can be represented descretely in a numerical grid. 

**Chapter 20.2 - Finite Difference Approximating Derivatives**

The finite derivative of a function f(x) at point a can be defined using the equation below:
$f'(a) = lim_{x->a}\frac{f(x)-f(a)}{x-a}$
This is useful in finding the slope at the given point by using values of the function of neighboring points.

Multiple different differences can be used when approximating the slope:
- The forward difference - $f'(x_j)=\frac{f(x_{j+1})-f(x_j)}{x_{j+1}-x_j}$
- The backward difference - $f'(x_j)=\frac{f(x_j)-f(x_{j-1})}{x_j-x_{j-1}}$
- The central difference - $f'(x_j)=\frac{f(x_{j+1})-f(x_{j-1})}{x_{j+1}-x_{j-1}}$

The approximation for the derivative of f can be solved with Taylor series, but O(h) can be added to create the forward difference equation above. O(h) defines the accuracy of the formula. The Taylor series can also derive the backward difference equation as well. The last equation, the central difference, can be defined by subtracting both formulas.Here, O(h) becomes O(h^2).

*Approximate the derivative of the function f(x) = x^2 using numerical differentiation:*

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# h will represent the step size of the
# grid and x will define the grid. y is
# our approximation function from above.

h = 0.1
x = np.arange(0, 2*np.pi, h) 
y = x**2

# The forward difference is calculated
# first then the exact solution is inputed.

forward_diff = np.diff(y)/h 
x_diff = x[:-1:] 
exact_solution = 2*(x_diff) 

# Lastly, the solutions are plotted
# against one another and the plot
# is labeled appropriately.

plt.style.use('seaborn-poster')
%matplotlib inline
plt.figure(figsize = (12, 8))
plt.grid('b')
plt.plot(x_diff, forward_diff, '--', \
         label = 'Finite Difference Approximation')
plt.plot(x_diff, exact_solution, \
         label = 'Exact Solution')
plt.title('Approximation Comparison')
plt.legend()
plt.show()

# Below the graph, the max error between
# numerical derivative and exact solution
# is calculated and printed.

max_error = max(abs(exact_solution - forward_diff))
print("The error between the two solutions is " + str(max_error))
```

*Approximate the derivative of the function f(x) = cos(x) using numerical differentiation:*

```python
h = 0.1
x = np.arange(0, 2*np.pi, h) 
y = np.cos(x)

forward_diff = np.diff(y)/h 
x_diff = x[:-1:] 
exact_solution = -1*np.sin(x_diff) 

plt.style.use('seaborn-poster')
%matplotlib inline
plt.figure(figsize = (12, 8))
plt.grid('b')
plt.plot(x_diff, forward_diff, '--', \
         label = 'Finite Difference Approximation')
plt.plot(x_diff, exact_solution, \
         label = 'Exact Solution')
plt.title('Approximation Comparison')
plt.legend()
plt.show()

max_error = max(abs(exact_solution - forward_diff))
print("The error between the two solutions is " + str(max_error))
```

Taylor series can also be used to approximate higher order derivatves like f'' etc. The equation for a second derivative approximation can be written as:
$f''(x_j) = \frac{f(x_{j+1})-2f(x_j)+f(x_{j-1})}{h^2}$

**Chapter 20.4 - Numerical Differentiation with Noise**

Data can be contaminated with noise which happens with the value is off by a small margin. Error in calculations or machines is the major cause of noise in real world problems.

*Compare the noise between the first derivative and second derivative of f(x) = sin(x) given omega = 100 and epsilon = 0.01:*

```python
x = np.arange(0, 2*np.pi, 0.01) 
omega = 100
epsilon = 0.01

# The functions for y and y with noise
# are computed with the defined variables
# above.

y = np.sin(x)
y_noise = y + epsilon*np.cos(omega*x)

# The graph of y and y with noise are
# plotted against eachother for comparison.

plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'sin(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'sin(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.grid('b')
plt.title('Noise in f(x)')
plt.legend()
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.01) 
y = np.cos(x) 
y_noise = y - epsilon*omega*np.sin(omega*x)

plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'Derivative sin(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'Derivative of sin(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.grid('b')
plt.title('Noise in Derivative f(x)')
plt.legend()
plt.show()
```
